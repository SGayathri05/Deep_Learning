## AND GATE USING KERAS(MULTILAYER PERCEPTRON)

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

input = np.array([[0,0],[0,1],[1,0],[1,1]])
output = np.array([[0],[0],[0],[1]])

model = Sequential()
model.add(Dense(16,input_dim=2, activation='relu'))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='rmsprop', metrics=['binary_accuracy'])

model.fit(input,output,epochs=100,verbose=1)

print(model.predict(input).round())

```
